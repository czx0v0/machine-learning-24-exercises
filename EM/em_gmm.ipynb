```python
import numpy as np
import time
import random
import math
```

```python
Y = np.array([-67,-48,6,8,14,16,23,24,28,29,41,49,56,60,75])
```

```python
# 辅助函数
def cal_gaussian_phi_k(y,sigma2_k,mu_k):
    res = 1/(math.sqrt(2*math.pi*sigma2_k))
    # print(res)
    res = res*math.exp(-((y-mu_k)*(y-mu_k))/(2*sigma2_k))
    # print(res)
    return res

# 
def gmm_em(Y,alpha_1,sigma2_1,sigma2_2,mu_1,mu_2,iter_max=5):
    print(f'初值：alpha_1 = {alpha_1},sigma2_1 = {sigma2_1},sigma2_2 = {sigma2_2},mu_1 = {mu_1},mu_2 = {mu_2}')
    N = len(Y)
    alpha_1i = alpha_1
    sigma2_1i = sigma2_1
    mu_1i = mu_1
    mu_2i = mu_2
    sigma2_2i = sigma2_2
    
    gamma_j1_N = np.ones(N)
    gamma_j2_N = np.ones(N)
    gamma_j2_yj_N = np.ones(N)
    gamma_j1_yj_N = np.ones(N)
    gamma_yj_mu1_2_N = np.ones(N)
    gamma_yj_mu2_2_N = np.ones(N)
    
    for i in range(iter_max):
        for j in range(N):
            yj = Y[j]
            # print(f'yj={yj}mu_1i = {mu_1i},mu_2i = {mu_2i},sigma2_1i = {sigma2_1i},sigma2_2i = {sigma2_2i},alpha_1i = {alpha_1i}')
            # print(cal_gaussian_phi_k(yj,sigma2_1i,mu_1i))
            # print(cal_gaussian_phi_k(yj,sigma2_2i,mu_2i))
            gamma_j1_N[j] = (alpha_1i*cal_gaussian_phi_k(yj,sigma2_1i,mu_1i))/((alpha_1i*cal_gaussian_phi_k(yj,sigma2_1i,mu_1i))+(1-alpha_1i)*cal_gaussian_phi_k(yj,sigma2_2i,mu_2i))
            gamma_j2_N[j] =  ((1-alpha_1i)*cal_gaussian_phi_k(yj,sigma2_2i,mu_2i))/((alpha_1i*cal_gaussian_phi_k(yj,sigma2_1i,mu_1i))+(1-alpha_1i)*cal_gaussian_phi_k(yj,sigma2_2i,mu_2i))
            gamma_j1_yj_N[j] = yj*gamma_j1_N[j]
            gamma_j2_yj_N[j] = yj*gamma_j2_N[j]
            gamma_yj_mu1_2_N[j] = gamma_j1_N[j]*((yj-mu_1i)**2)
            gamma_yj_mu2_2_N[j] = gamma_j1_N[j]*((yj-mu_2i)**2)
            
        # M步
        mu_1i_new = sum(gamma_j1_yj_N)/sum(gamma_j1_N)
        mu_2i_new = sum(gamma_j2_yj_N)/sum(gamma_j2_N)
        sigma2_1i_new = sum(gamma_yj_mu1_2_N)/sum(gamma_j1_N)
        sigma2_2i_new = sum(gamma_yj_mu2_2_N)/sum(gamma_j2_N)
        alpha_1i_new = sum(gamma_j1_N)/N
        mu_1i = mu_1i_new
        mu_2i = mu_2i_new
        sigma2_1i = sigma2_1i_new
        sigma2_2i = sigma2_2i_new
        alpha_1i = alpha_1i_new
        
        print(f'第{i+1}轮：mu_1i = {mu_1i},mu_2i = {mu_2i},sigma2_1i = {sigma2_1i},sigma2_2i = {sigma2_2i},alpha_1i = {alpha_1i}')
        
        
    return alpha_1i,sigma2_1i,sigma2_2i,mu_1i,mu_2i

        
        
        
        
```

```python
alpha_1i,sigma2_1i,sigma2_2i,mu_1i,mu_2i = gmm_em(Y,0.5,100,100,10,10,100)
print(f'mu_1i = {mu_1i},mu_2i = {mu_2i},sigma2_1i = {sigma2_1i},sigma2_2i = {sigma2_2i},alpha_1i = {alpha_1i}')
```
