```python
import random
import numpy as np
import time

def em_3coin(pi,p,q,n):
    res = []
    res_a = []
    res_b = []
    res_c = []
    for i in range(n):
        #先硬币A
        if random.random() < pi:
            coin= 'B'
            p_head = p
        else:
            coin = 'C'
            p_head = q
        if random.random() < p_head:
            res.append(1)
        else:
            res.append(0)
        if coin=='B':
            if random.random() < p:
                res_b.append(1)
            else:
                res_b.append(0)
            res_a.append(1)
        else:
            if random.random() < q:
                res_c.append(1)
            else:
                res_c.append(0)
            res_a.append(0)
    p_a = sum(res_a)/len(res_a)
    p_b = sum(res_b)/len(res_b)
    p_c = sum(res_c)/len(res_c)
    
    return res,p_a,p_b,p_c


            
            
```

```python
pi = 0.2
p = 0.3
q = 0.8
n = 10000
s = time.time()
print(f's= {time.time()}')
Y,pi,p,q = em_3coin(pi,p,q,n)
Y = np.array(Y)
print(f'time= {time.time()-s}')
print(f'pi= {pi},p={p},q= {q},n= {n},all Y =1 P:{sum(Y)/len(Y)} ')
```

```python
# EM 参数反推
pi_0= 0.46
p_0= 0.55
q_0= 0.67
episodes = 100
count = 1
while count <= episodes:
    mu = pi_0*p_0**Y * (1-p_0)**(1-Y)/(pi_0*p_0**Y*(1-p_0)**(1-Y)+(1-pi_0)*q_0**Y * (1-q_0)**(1-Y))
    pi = (1/n)*sum(mu)
    p = sum(Y*mu)/sum(mu)
    q = sum((1-mu)*Y)/sum(1-mu)
    print(f'count= {count}次迭代，pi={pi},p={p},q= {q}')
    pi_0= pi
    p_0= p
    q_0= q
    count +=1
    
s = time.time()
print(f'pi={pi},p={p},q={q}')
Y,pi,p,q = em_3coin(pi,p,q,n)
Y = np.array(Y)
print(f'time: {time.time()-s}')
print(f'pi= {pi},p={p},q= {q},all Y =1 P:{sum(Y)/len(Y)} ')
        
```
