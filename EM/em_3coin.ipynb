```python
import numpy as np
import time
```

```python
Y = np.array([1,1,0,1,0,0,1,0,1,1])
```

```python
def cal_em_3coins(Y,pi_0 =0.46 ,p_0= 0.55,q_0= 0.67,iter_max = 20):
    # 初值
    pi_i= pi_0
    p_i= p_0
    q_i= q_0
    # 迭代步骤
    n = len(Y)
    # 存储j方便计算
    mu = np.zeros(n) # mu_j
    y_times_mu = np.zeros(n) # y_j*mu_j
    y_times_1_mu = np.zeros(n) # y_j*(1-mu_j)
    mu_1 = np.zeros(n) # (1-mu_j)

    for i in range(iter_max):
        # E步
        for j in range(n):
            mu[j] = (pi_i*p_i**Y[j] * (1-p_i)**(1-Y[j]))/(pi_i*p_i**Y[j] * (1-p_i)**(1-Y[j])+(1-pi_i)*q_i**Y[j] * (1-q_i)**(1-Y[j]))
            y_times_mu[j] = mu[j]*Y[j]
            y_times_1_mu[j] = Y[j]*(1-mu[j])
            mu_1[j] = 1-mu[j]
        # M步
        pi_i_n = (1/n)*sum(mu)
        p_i_n = sum(y_times_mu)/sum(mu)
        q_i_n = sum(y_times_1_mu)/sum(mu_1)
        pi_i  = pi_i_n
        p_i = p_i_n
        q_i = q_i_n
        print(f'第{i+1}轮迭代，y_j = 1,mu={mu[0]}.y_j=0,mu={mu[2]}')
        print(f'第{i+1}轮迭代，pi_{i+1}={pi_i},p_{i+1}={p_i},q_{i+1}= {q_i}')
    return pi_i, p_i, q_i
```

```python

pi,p,q = cal_em_3coins(Y,0.5,0.5,0.5,30)
print(f'pi= {pi},p={p},q= {q}')
```
