```python
import numpy as np
from scipy.linalg import null_space
```

```python
A = np.array([[2,4],[1,3],[0,0],[0,0],[0,0]])
# A =np.array([[4,0],[3,0],[0,0]])
m,n = A.shape
if m<n:
    A = A.conj().T
print(m,n)
print(A)
```

```python
ATA = np.array(A.conj().T@A)
print(ATA)
```

```python
ATA_value,ATA_vector = np.linalg.eig(ATA)
print(ATA_vector)
print(ATA_value)
```

```python
sigma_list = []
argsort_value = np.argsort(-ATA_value)
ATA_sort_value = ATA_value[argsort_value]
print(ATA_sort_value)
for i in range (0,ATA_sort_value.shape[0]):
    value = np.sqrt(ATA_sort_value[i])
    sigma_list.append(value)
    print(value)
print(sigma_list)

    
```

```python
sigma_matrix = np.zeros((m,n))
sigma_matrix[:len(sigma_list),:len(sigma_list)] = np.diag(sigma_list)
print(sigma_matrix)
```

```python

VTMatrix = ATA_vector[argsort_value]
print(VTMatrix)
```

```python
U = np.zeros((m,m))
b = np.zeros((1,n))
print(A.T)
for i in range(m):
    
    if i < len(sigma_list) and (sigma_list[i]>0):
        uj = 1/sigma_list[i] * (A@ VTMatrix[i])
        print(uj)
        U[i] = uj
    else:
        # mr,_ = np.linalg.qr(A.T)
        mr = np.linalg.lstsq(A.T,b.T,rcond=None)
        print(mr[0].T)
        U[i] = mr[0][0].T
        
print(U)
```

```python
print(U.T@sigma_matrix@VTMatrix)
```

```python
AAT_value,AAT_vector = np.linalg.eig(AAT)
print(AAT_vector)
AAT_value[AAT_value<1e-5] = 0
print(AAT_value)
argsort2_value = np.argsort(-AAT_value)
print(argsort2_value)
```
